```python
from io import StringIO

s3 = boto3.client('s3')
bucket_name = 'cmse492-lesurek1-nyc311-339713112540-us-east-1-an' # replace with ame (make sure you have the account regional suffix)your bucket n
file_name = 'modeling_data.csv' # make sure to include the path if it's in a folder, e.g. 'modeling/modeling_data.csv'

obj = s3.get_object(Bucket=bucket_name, Key=file_name)
data = obj['Body'].read().decode('utf-8')
df = pd.read_csv(StringIO(data))
```

```python
# First few rows
print("First few rows:")
print(df.head())

# Data types and missing values
print("\n" + "="*50)
print("Data types and missing values:")
print(df.info())
```

```python
print("Summary statistics:")
print(df.describe())
```

```python
print("Target distribution (problems):")
print(df['problem'].value_counts().sort_index())

print("\nTarget distribution (%)")
print((df['problem'].value_counts() / len(df) * 100).sort_index().round(2))
```
